```python
from youtube_transcript_api import YouTubeTranscriptApi

# Replace 'VIDEO_ID' with the actual video ID
video_id = 'M_KHcplECJo'  # Example video ID; change as needed
transcript = YouTubeTranscriptApi.get_transcript(video_id)

# Format transcript output
full_text = "\n".join([segment['text'] for segment in transcript])
print(full_text)
```
